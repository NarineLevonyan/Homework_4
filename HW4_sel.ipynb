## Get the data using Selenium

```python
from selenium import webdriver
import time
from pprint import pprint
import re
```

```python
url = "https://www.youtube.com/feed/trending"
browser = webdriver.Chrome()
browser.get(url)
```

```python
#browser.page_source
```

```python
#title
```

```python
for i in browser.find_elements_by_xpath("//a[@class='yt-simple-endpoint style-scope ytd-video-renderer']"):
    print(i.get_attribute("innerHTML").replace("\n", ""))
```

```python
#href
```

```python
for i in browser.find_elements_by_xpath("//a[@class='yt-simple-endpoint style-scope ytd-video-renderer']"):
    print(i.get_attribute("href"))
```

```python
views =[i.get_attribute("aria-label") for i in browser.find_elements_by_xpath("//a[@id='video-title']")]
view = [re.findall("([0-9]+)\sviews", i.replace(",", "")) for i in views]
```

```python
import itertools
merged_views = list(itertools.chain(*view))
pprint(merged_views)
```
