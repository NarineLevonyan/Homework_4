## Get the data using xml

```python
import requests
from lxml import html
from lxml.etree import tostring
from pprint import pprint
```

```python
base_url = "https://www.youtube.com"
url = "https://www.youtube.com/feed/trending"
response = requests.get(url)
page = response.content
tree = html.document_fromstring(page) #html to xml
```

```python
tostring(tree)
```

```python
#URL
urls = [base_url + i for i in tree.xpath("//a[@class='yt-uix-tile-link yt-ui-ellipsis yt-ui-ellipsis-2 yt-uix-sessionlink      spf-link ']/@href")]
del(urls[4:9])
pprint(urls)
```

```python
#duration
duration = [i.text_content().replace(" - Duration: ", '').replace(".", "") for i in tree.xpath("//span[@class='accessible-description']")]
del(duration[4:9])
pprint(duration)
```

```python
#view
views = [i.text_content().replace('views', '') for i in tree.xpath("//ul[@class='yt-lockup-meta-info']/li[2]")]
del(views[4:9])
pprint(views)
```

```python
#titles
titles = tree.xpath("//a[@class='yt-uix-tile-link yt-ui-ellipsis yt-ui-ellipsis-2 yt-uix-sessionlink      spf-link ']/@title")
del(titles[4:9])
pprint(titles)
```

```python
#username
usernames = [i.text_content() for i in tree.xpath("//a[@class='yt-uix-sessionlink       spf-link ']")]
del(usernames[4:9])
pprint(usernames)
```
