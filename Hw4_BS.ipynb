```python
import requests  # get html
from bs4 import BeautifulSoup  # for scraping
from pprint import pprint  # pretty printing
import json
from dicttoxml import dicttoxml 
import numpy as np

url_trending = "https://www.youtube.com/feed/trending"
youtube_url = "https://www.youtube.com"


def get_info(url):
    response = requests.get(url)
    page = response.content
    page = BeautifulSoup(page, "html.parser")
    items = page.find_all("div", class_="yt-lockup-content")
    youtube_items = []
    for i in items:
        a_tag = i.find("a", class_="yt-uix-tile-link")
        if a_tag is not None:
            youtube_items.append({
                "url": youtube_url + a_tag.get("href"),
                "title": a_tag.get("title"),
                "duration": i.find("span", class_="accessible-description").get_text().replace(" - Duration: ", ""),
                "views": i.find("ul", class_="yt-lockup-meta-info").find_all("li")[1].get_text().replace(" views", ""),
                "username" : i.find("div", class_="yt-lockup-byline ").find("a").get_text()
            })
    return youtube_items

pprint(get_info(url_trending))
```

```python
data = (get_info(url_trending))

with open("youtube_trending.json", "w") as f:
    json.dump(data, f, indent=5) 
```

```python
import pandas as pd

df = pd.DataFrame(data)
print(df)

df.to_csv("youtube_trending.csv")
```

```python
youtube_trending_xml = dicttoxml(data)
youtube_trending_xml = youtube_trending_xml.decode("utf-8")
with open("youtube_trending.xml", "w", encoding='utf-8') as f:
    f.write(youtube_trending_xml )
```
